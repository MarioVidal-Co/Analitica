```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

from google.colab import files
from google.colab import drive
drive.mount('/content/drive')
```

```python
nxl='/content/drive/MyDrive/AnaÃÅlitica de Negocios/2. BD1_Travel_Data_int.xlsx'
XDB=pd.read_excel(nxl,sheet_name=1)
XDB=XDB.dropna()
XDB.head(100)

XD=np.array(XDB[['College_B','FoodSpend','Income','TravelSpend']])
print(XD)
yd=np.array(XDB[['Credit_Card']])
print(yd)
```

```python
mar=DecisionTreeClassifier(criterion='gini',max_depth=4)
mar.fit(XD,yd)
ydp=mar.predict(XD)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(yd,ydp)
print(cm)

VN=cm[0,0]
FP=cm[0,1]
FN=cm[1,0]
VP=cm[1,1]

Ex=(VN+VP)/(VN+VP+FN+FP)
print("La Exactitud alcanzada por el modelo es:",Ex)
Se= (VP)/(FN+VP)
print("La sensibilidad alcanzada por el modelo es:",Se)
Es= (VN)/(VN+FP)
print("La Especificidad alcanzada por el modelo es:",Es)
```

```python
from sklearn.tree import export_graphviz
from pydotplus import graph_from_dot_data

vs=['College_B','FoodSpend','Income','TravelSpend']
dot_graph=export_graphviz(mar,feature_names=vs)
graph=graph_from_dot_data(dot_graph)
graph.write_png('arbol2.png')
```

```python
XDB2=pd.read_excel(nxl,sheet_name=1)
XDB2=XDB2.dropna()
XDB2.head(100)

XD2=np.array(XDB2[['College_B','FoodSpend','Income','TravelSpend']]) #Datos de entrada
print(XD2)

#Pronosticamos
ydp2=mar.predict(XD2)
print(ydp2)
print("El n√∫mero total de personas que usar√≠an su TC es:",np.sum(ydp2))
print("El n√∫mero total de datos es:",len(ydp2))
print("El n√∫mero de personas que usar√≠an su TC es:",len(ydp2)-np.sum(ydp2))
```
